# Consulta 42
Si la experiencia de un usuario es la cantidad de logs en los que participó, ¿cuál es
la tasa de contenidos cuya última revisión no tiene comentario en función de la
experiencia de su revisor?

```python
logs = pd.read_csv('data/logs.csv', usecols=['contributor_id', 'contributor_ip'], low_memory = False)
logs = logs.loc[logs['contributor_id'].notna() | logs['contributor_ip'].notna() ] #me quedo con los que tienen id o ip
logs
```

```python
#calculo la experiencia de cada usuario
experiencias_ID = logs['contributor_id'].value_counts().to_frame('experiencia')
experiencias_IP = logs['contributor_ip'].value_counts().to_frame('experiencia')
del logs
```

```python
revisiones = pd.read_csv('data/contents.csv', usecols = ['revisor_id', 'revisor_ip', 'revisor_comment'])
revisiones
```

```python
#me quedo solo con las experiencias de los usuarios que revisaron contenido
experiencias_ID = experiencias_ID.loc[experiencias_ID.index.isin(revisiones['revisor_id'])]
experiencias_IP = experiencias_IP.loc[experiencias_IP.index.isin(revisiones['revisor_ip'])]
experiencias_ID.index = experiencias_ID.index.astype('int64')
```

```python
#guardo la cantidad de revisiones por cada revisor
cant_revisionesID = revisiones['revisor_id'].value_counts().to_frame('cant_revisiones')
cant_revisionesIP = revisiones['revisor_ip'].value_counts().to_frame('cant_revisiones')
```

```python
#me quedo solo con las revisiones sin comentario
revisiones = revisiones.loc[revisiones['revisor_comment'].isna()]
cant_revisionesID['cant_revisiones_sinCom'] = revisiones['revisor_id'].value_counts()
cant_revisionesIP['cant_revisiones_sinCom'] = revisiones['revisor_ip'].value_counts()
```

```python
cant_revisionesID.index = cant_revisionesID.index.astype('int64')
cant_revisionesID.fillna(axis = 0, inplace = True, value = {'cant_revisiones_sinCom' : 0})
cant_revisionesID['cant_revisiones_sinCom'] = cant_revisionesID['cant_revisiones_sinCom'].astype('int64')
cant_revisionesID
```

```python
cant_revisionesIP.fillna(axis = 0, inplace = True, value = {'cant_revisiones_sinCom' : 0})
cant_revisionesIP['cant_revisiones_sinCom'] = cant_revisionesIP['cant_revisiones_sinCom'].astype('int64')
cant_revisionesIP
```

```python
#calculo el porcentaje de revisiones sin comentarios por revisor
cant_revisionesID['% sin comentario'] = cant_revisionesID['cant_revisiones_sinCom'] / cant_revisionesID['cant_revisiones'] * 100
cant_revisionesIP['% sin comentario'] = cant_revisionesIP['cant_revisiones_sinCom'] / cant_revisionesIP['cant_revisiones'] * 100
```

```python
cant_revisionesID
```

```python
experiencias_ID = experiencias_ID.join(cant_revisionesID['% sin comentario'], how='inner')
experiencias_IP = experiencias_IP.join(cant_revisionesIP['% sin comentario'], how='inner')
```

```python
experiencias_ID
```

```python
# Tasa de contenidos cuya última revisión no tiene comentario en función de la experiencia de su revisor por ID
experiencias_ID.groupby(by='experiencia', axis=0).mean()
```

```python
# Tasa de contenidos cuya última revisión no tiene comentario en función de la experiencia de su revisor por IP
experiencias_IP.groupby(by='experiencia', axis=0).mean()
```

# Consulta 59
¿Cuál es el idioma para el cual sus usuarios realizan más agradecimientos en
promedio? ¿Y el de menos agradecimientos? Calcule lo mismo para quienes reciben
agradecimientos.

```python
logs = pd.read_csv('data/logs.csv', usecols = ['contributor_username', 'title', 'contributor_id', 'action'])
lenguajes = pd.read_csv('data/languages.csv', usecols = ['babel_user', 'babel_lang'])
```

```python
agradecimientos = logs.loc[logs['action'] == 'thank' ]
del logs
agradecimientos.drop(columns = 'action', inplace = True)
agradecimientos
```

```python
cantAgradecimientos = agradecimientos['contributor_id'].value_counts().rename('cantAgradecimientos')
cantAgradecimientos
```

```python
lenguajes = lenguajes.set_index('babel_user').join(cantAgradecimientos, how= 'inner')
lenguajes
```

```python
# lenguajes para los cuales sus usuarios realizan más agradecimientos en promedio
lenguajes.groupby(by='babel_lang').mean().nlargest(5, columns= 'cantAgradecimientos')
```

```python
# lenguajes para los cuales sus usuarios realizan menos agradecimientos en promedio
lenguajes.groupby(by='babel_lang').mean().nsmallest(5, columns= 'cantAgradecimientos')
```

```python
# calculo lo mismo pero para los agradecimientos recibidos
agradecimientos
```

```python
remitentes_agra = agradecimientos['title'].apply(lambda linea : (str(linea).split(':', 1)))
remitentes_agra
```

```python
# me quedo con los nombres de usuario
remitentes_agra = remitentes_agra.loc[remitentes_agra.apply(lambda linea: len(linea) > 1)]
remitentes_agra = remitentes_agra.apply(lambda linea: linea[1])
```

```python
remitentes_agra
```

```python
remitentes_agra = remitentes_agra.value_counts().to_frame('cantAgraRecibidos')
```

```python
remitentes_agra
```

```python
users = pd.read_csv('data/logs.csv', usecols = ['contributor_username', 'contributor_id'])
users = users.set_index('contributor_username')
users.drop_duplicates(inplace=True)
users
```

```python
remitentes_agra = remitentes_agra.join(users, how= 'inner')
del users
remitentes_agra
```

```python
remitentes_agra.set_index('contributor_id', inplace=True)
```

```python
lenguajes = pd.read_csv('data/languages.csv', usecols = ['babel_user', 'babel_lang'])
lenguajes.set_index('babel_user', inplace=True)
```

```python
lenguajes = lenguajes.join(remitentes_agra,how='inner')
```

```python
# lenguajes para los cuales sus usuarios reciben más agradecimientos en promedio
lenguajes.groupby(by='babel_lang', axis=0).mean().nlargest(5, columns='cantAgraRecibidos')
```

```python
# lenguajes para los cuales sus usuarios reciben menos agradecimientos en promedio
lenguajes.groupby(by='babel_lang', axis=0).mean().nsmallest(5, columns='cantAgraRecibidos')
```
